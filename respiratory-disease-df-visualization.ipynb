###### -- Download the Dataset.xlsx file to DataFrame.
The &quot;Diagnosis&quot; column has three variants of diagnosis: COPD – chronic obstructive pulmonary disease, asthma – bronchial asthma and pneumonia.

The table also contains some binary features encoded 0 and 1, several rank features and quantitative features.

```python
# importing libraries to work with dataframe (pandas)
import pandas as pd

# reading an Excel file
df = pd.read_excel('data/Dataset.xlsx')
df.set_index('Case history number', inplace=True)
df
```

###### -- Display the description in a separate DataFrame. This will provide information about all symptoms.

```python
description_df = df.describe()
description_df
```

```python
# counting the missing values across the DataFrame
missing_values = df.isnull().sum()

# sorting values in the DataFrame so that columns with more missing values will be on top
missing_values = missing_values.sort_values(ascending=False)

# displaying only columns with more than 0 missing values
missing_values[missing_values > 0]
```

###### -- Delete all features that contain missing values, which are marked in the table as NaN (Not a Number) or those patients that contain too many missing values (for example, a patient with ID 995)

```python
# deleting all the features (columns) that contain missing values
# using the method df.drop() - remove the specified row or column
df_col_wo_nan1 = df.drop(df.columns[df.isna().any()], axis=1)
df_col_wo_nan1
```

```python
# deleting all the features (columns) that contain missing values
# using the method df.dropna() - drop the rows and columns with NaN values
df_col_wo_nan2 = df.dropna(axis=1)
df_col_wo_nan2
```

```python
# deleting all the patients that contain too many missing values
# let's assume that too many is 80%
# using the method df.drop() - remove the specified row or column
df_row_wo_nan1 = df[df.isna().mean(axis=1) <= 0.2]
df_row_wo_nan1
```

```python
# let's assume that too many is more than 2 NaN values
df_row_wo_nan2 = df.dropna(thresh=df.shape[1] - 2)
df_row_wo_nan2
```

###### -- Create separate tables for women and men.

```python
genders = df_col_wo_nan1.groupby('gender')

# separating tables for women and men assuming that 0 corresponds to women and 1 to man
women_df = genders.get_group(0)
men_df = genders.get_group(1)
```

```python
women_df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# let's visualize the percentage of diagnoses across women in bar and pie chart
w_diagnosis = women_df['Diagnosis'].value_counts()
w_diagnosis = w_diagnosis.sort_values(ascending=False)

# plotting the bar chart
plt.figure(figsize=(8, 6))
w_diagnosis_plot = sns.barplot(x=w_diagnosis.index, y=w_diagnosis.values)
plt.xlabel('Diagnoses')
plt.ylabel('Number of patients')
plt.title('Total number of women across diagnoses')

# annotating the number of women patients with certain diagnoses on top of each bar
for p in w_diagnosis_plot.patches:
    w_diagnosis_plot.annotate(f'{int(p.get_height())}', (p.get_x() + p.get_width() / 2, p.get_height()), ha='center', va='bottom')
plt.show()

# plotting the pie chart
plt.figure(figsize=(6, 6))
plt.pie(w_diagnosis, labels=w_diagnosis.index, autopct='%1.1f%%', startangle=90)
plt.title('Distribution of diagnoses for women')
plt.show()
```

```python
men_df
```

```python
# let's visualize the percentage of diagnoses across men in bar and pie chart
m_diagnosis = men_df['Diagnosis'].value_counts()
m_diagnosis = m_diagnosis.sort_values(ascending=False)

# plotting the bar chart
plt.figure(figsize=(8, 6))
m_diagnosis_plot = sns.barplot(x=m_diagnosis.index, y=m_diagnosis.values)
plt.xlabel('Diagnoses')
plt.ylabel('Number of patients')
plt.title('Total number of men across diagnoses')

# annotating the number of women patients with certain diagnoses on top of each bar
for p in m_diagnosis_plot.patches:
    m_diagnosis_plot.annotate(f'{int(p.get_height())}', (p.get_x() + p.get_width() / 2, p.get_height()), ha='center', va='bottom')
plt.show()

# plotting the pie chart
plt.figure(figsize=(6, 6))
plt.pie(m_diagnosis, labels=m_diagnosis.index, autopct='%1.1f%%', startangle=90)
plt.title('Distribution of diagnoses for men')
plt.show()
```

```python
# let's visualize the distribution of diagnoses across women and men in stacked bar chart
import altair as alt

# combining dataframes for visualization
combined_df = pd.concat([women_df, men_df])

# creating a stacked bar chart
bars = alt.Chart(combined_df).mark_bar().encode(
    y=alt.Y("Diagnosis:N", axis=alt.Axis(title="Diagnosis")),
    x=alt.X("count():Q", axis=alt.Axis(title="Number of patients")),
    # naming genders in legend instead of numbers
    color=alt.Color('gender:N', legend=alt.Legend(title='Gender', labelExpr="datum.value == 0 ? 'Women' : 'Men'")),
    order=alt.Order("gender:N", sort="descending")
)

bars.properties(
    height=300,
    width=700,
    title='Distribution of diagnoses for men and women'
)
```

###### -- Calculate the total number of patients with each type of disease and their percentage in the total number of patients

```python
# calculating total num of patiens with each type of disease in the initial dataframe
diagnosis = df['Diagnosis'].value_counts()

# calculating percentage of diagnosis in the total number of patients
total_patients = len(df)
percentage = (diagnosis / total_patients) * 100

# displaying info in dataframe
diagnosis_df = pd.DataFrame({
    'Diagnosis': diagnosis.index,
    'Total patients': diagnosis.values,
    'Percentage': percentage.values
})

diagnosis_df = diagnosis_df.sort_values('Total patients', ascending=False)
diagnosis_df
```

```python
# let's visualize the number results with a bar chart
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))
diagnosis_plot = sns.barplot(x="Diagnosis", y="Total patients", data=diagnosis_df)
plt.xlabel('Diagnoses')
plt.ylabel('Number of patients')
plt.title('Total number of patiens with each diagnosis')

# annotating the number of patiens with certain diagnoses on top of each bar
for p in diagnosis_plot.patches:
    diagnosis_plot.annotate(f'{int(p.get_height())}', (p.get_x() + p.get_width() / 2, p.get_height()), ha='center', va='bottom')
plt.show()
```

```python
# let's visualize the percentage results with a pie chart
plt.figure(figsize=(6, 6))
plt.pie(diagnosis_df['Total patients'], labels=diagnosis_df['Diagnosis'], autopct='%1.1f%%', startangle=90)
plt.axis('equal')
plt.title('Diagnosis percentage in the total number of patients')
plt.show()
```

###### -- Apply a method for data preprocessing (standardizing or normalizing).

```python
# let's assign numbers 1-3 to the corresponding diagnoses
encoding = {"Pneumonia": 0, "COPD": 1, "Asthma": 2}
df_col_wo_nan1['Diagnosis'] = df_col_wo_nan1['Diagnosis'].replace(encoding)
df_col_wo_nan1
```

```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler

# preparing table for normalization - removing column 'Diagnosis with categorical data'
normalized_df = df_col_wo_nan1.drop('Diagnosis', axis=1).copy()
df_diagnosis = df_col_wo_nan1.reset_index().drop('Case history number', axis=1).copy()

# useful if relative difference is more important than absolute value
# normalizing data
scaler = MinMaxScaler()
normalized_df = pd.DataFrame(scaler.fit_transform(normalized_df), columns=normalized_df.columns)
normalized_df
```

```python
# useful if data is of different units
# standardizing data
scaler_standard = StandardScaler()
standard_df = df_col_wo_nan1.drop('Diagnosis', axis=1).copy()
standard_df = pd.DataFrame(scaler_standard.fit_transform(standard_df), columns=standard_df.columns)
standard_df
```

##### -- Create new DataFrame that contains 3 principal components and new DataFrame that contains 2 TSNE features.

###### - fit principal components analysis method to preprocessed DataFrame without Diagnosis column;

```python
from sklearn.decomposition import PCA

# initializing pca with 3 components
pca = PCA(n_components=3)

# fitting pca to the preprocessed data
principal_comp = pca.fit_transform(normalized_df)
principal_comp
```

###### - create new DataFrame that contains 2 principal components and a column with diagnosis;

```python
pca_df = pd.DataFrame(data=principal_comp[:, :2], 
                      columns=['Principal component 1',
                               'Principal component 2'])
pca_df['Diagnosis'] = df_diagnosis['Diagnosis']
pca_df
```

```python
# visualizing the result
fig = plt.figure(figsize = (6,6))
ax = fig.add_subplot(1,1,1) 
ax.set_xlabel('Principal Component 1', fontsize = 15)
ax.set_ylabel('Principal Component 2', fontsize = 15)
ax.set_title('PCA 2D', fontsize = 20)
diagnoses = [0, 1, 2]
colors = ['r', 'g', 'b']
for diagnosis, color in zip(diagnoses,colors):
    indicesToKeep = pca_df['Diagnosis'] == diagnosis
    ax.scatter(pca_df.loc[indicesToKeep, 'Principal component 1']
               , pca_df.loc[indicesToKeep, 'Principal component 2']
               , c = color
               , s = 50)
ax.legend(diagnoses)
ax.grid()
```

###### - create new DataFrame that contains 3 principal components and a column with diagnosis;

```python
pca_df_3d = pd.DataFrame(data=principal_comp, 
                         columns=['PC1', 
                                  'PC2', 
                                  'PC3'])
pca_df_3d['Diagnosis'] = df_diagnosis['Diagnosis']
pca_df_3d
```

```python
# visualizing the result
fig = plt.figure(figsize = (7,7))
ax = fig.add_subplot(111, projection='3d')
ax.set_xlabel('Principal Component 1', fontsize = 10)
ax.set_ylabel('Principal Component 2', fontsize = 10)
ax.set_zlabel('Principal Component 3', fontsize = 10)
ax.set_title('PCA 3D', fontsize = 15)
diagnoses = [0, 1, 2]
colors = ['r', 'g', 'b']
for diagnosis, color in zip(diagnoses,colors):
    indicesToKeep = pca_df_3d['Diagnosis'] == diagnosis
    ax.scatter(pca_df_3d.loc[indicesToKeep, 'PC1'],
               pca_df_3d.loc[indicesToKeep, 'PC2'],
               pca_df_3d.loc[indicesToKeep, 'PC3'],
               c = color, alpha = 1, edgecolor='k',
               s = 50)
ax.view_init(elev=-150, azim=70)
ax.legend(diagnoses)
ax.grid()
```

###### - fit TSNE method to preprocessed DataFrame without Diagnosis column;

```python
from sklearn.manifold import TSNE

# initializing t-SNE with 2 components
tsne = TSNE(n_components=2)

# fitting t-SNE to the preprocessed data
tsne_result = tsne.fit_transform(normalized_df)
tsne_result
```

###### - create new DataFrame that contains 2 TSNE features and a column with diagnosis;

```python
tsne_df_2d = pd.DataFrame(data=tsne_result, 
                          columns=['t-SNE1',
                                   't-SNE2'])
tsne_df_2d['Diagnosis'] = df_diagnosis['Diagnosis']
tsne_df_2d
```

```python
# visualizing the result
fig = plt.figure(figsize = (6,6))
ax = fig.add_subplot(1,1,1) 
ax.set_xlabel('t-SNE Feature 1', fontsize = 15)
ax.set_ylabel('t-SNE Feature 2', fontsize = 15)
ax.set_title('2D t-SNE', fontsize = 20)
diagnoses = [0, 1, 2]
colors = ['r', 'g', 'b']
for diagnosis, color in zip(diagnoses,colors):
    indicesToKeep = tsne_df_2d['Diagnosis'] == diagnosis
    ax.scatter(tsne_df_2d.loc[indicesToKeep, 't-SNE1']
               , tsne_df_2d.loc[indicesToKeep, 't-SNE2']
               , c = color
               , s = 50)
ax.legend(diagnoses)
ax.grid()
```

```python
# initializing t-SNE with 3 components
tsne = TSNE(n_components=3)

# fitting t-SNE to the preprocessed data
tsne_result_3d = tsne.fit_transform(normalized_df)
tsne_result_3d
```

```python
tsne_df_3d = pd.DataFrame(data=tsne_result_3d, 
                         columns=['t-SNE1', 
                                  't-SNE2', 
                                  't-SNE3'])
tsne_df_3d['Diagnosis'] = df_diagnosis['Diagnosis']
tsne_df_3d
```

```python
# visualizing the result
fig = plt.figure(figsize = (7,7))
ax = fig.add_subplot(111, projection='3d')
ax.set_xlabel('t-SNE Feature 1', fontsize = 10)
ax.set_ylabel('t-SNE Feature 2', fontsize = 10)
ax.set_zlabel('t-SNE Feature 3', fontsize = 10)
ax.set_title('3D t-SNE', fontsize = 15)
diagnoses = [0, 1, 2]
colors = ['r', 'g', 'b']
for diagnosis, color in zip(diagnoses,colors):
    indicesToKeep = tsne_df_3d['Diagnosis'] == diagnosis
    ax.scatter(tsne_df_3d.loc[indicesToKeep, 't-SNE1'],
               tsne_df_3d.loc[indicesToKeep, 't-SNE2'],
               tsne_df_3d.loc[indicesToKeep, 't-SNE3'],
               c = color, alpha = 1, edgecolor='k',
               s = 50)
ax.view_init(elev=-150, azim=70)
ax.legend(diagnoses)
ax.grid()
```

##### -- Find better parameters for visualizing DataFrame by TSNE method (perplexity, n_iter, learning_rate).

```python
# tuning parameters for visualizing DataFrame by TSNE method 
# let's try to make distinct clusters within the group
tsne = TSNE(n_components=2, perplexity=15, learning_rate=400, n_iter=500, random_state=42)
tsne_components = tsne.fit_transform(normalized_df)

df_tsne_tuned = pd.DataFrame({
    'Diagnosis': df_diagnosis['Diagnosis'],
    't-SNE1': tsne_components[:, 0],
    't-SNE2': tsne_components[:, 1]
})
df_tsne_tuned
```

```python
# visualizing the result
fig = plt.figure(figsize = (6,6))
ax = fig.add_subplot(1,1,1) 
ax.set_xlabel('t-SNE Feature 1', fontsize = 15)
ax.set_ylabel('t-SNE Feature 2', fontsize = 15)
ax.set_title('2D t-SNE', fontsize = 20)
diagnoses = [0, 1, 2]
colors = ['r', 'g', 'b']
for diagnosis, color in zip(diagnoses,colors):
    indicesToKeep = df_tsne_tuned['Diagnosis'] == diagnosis
    ax.scatter(df_tsne_tuned.loc[indicesToKeep, 't-SNE1']
               , df_tsne_tuned.loc[indicesToKeep, 't-SNE2']
               , c = color
               , s = 50)
ax.legend(diagnoses)
ax.grid()
```

As we can see on the plot similar features are now closer to each other and at least form one group.

```python
# initializing t-SNE with 3 components and tuned parameters
tsne = TSNE(n_components=3, perplexity=5, learning_rate=10, n_iter=500)
#tsne = TSNE(n_components=3, perplexity=5, learning_rate=10, n_iter=500, random_state=42)

# fitting t-SNE to the preprocessed data
tuned_tsne_result_3d = tsne.fit_transform(normalized_df)

df_tsne_tuned_3d = pd.DataFrame(data=tuned_tsne_result_3d, 
                         columns=['t-SNE1', 
                                  't-SNE2', 
                                  't-SNE3'])
df_tsne_tuned_3d['Diagnosis'] = df_diagnosis['Diagnosis']
df_tsne_tuned_3d
```

```python
# visualizing the result
fig = plt.figure(figsize = (7,7))
ax = fig.add_subplot(111, projection='3d')
ax.set_xlabel('t-SNE Feature 1', fontsize = 10)
ax.set_ylabel('t-SNE Feature 2', fontsize = 10)
ax.set_zlabel('t-SNE Feature 3', fontsize = 10)
ax.set_title('3D t-SNE', fontsize = 15)
diagnoses = [0, 1, 2]
colors = ['r', 'g', 'b']
for diagnosis, color in zip(diagnoses,colors):
    indicesToKeep = df_tsne_tuned_3d['Diagnosis'] == diagnosis
    ax.scatter(df_tsne_tuned_3d.loc[indicesToKeep, 't-SNE1'],
               df_tsne_tuned_3d.loc[indicesToKeep, 't-SNE2'],
               df_tsne_tuned_3d.loc[indicesToKeep, 't-SNE3'],
               c = color, alpha = 1, edgecolor='k',
               s = 50)
ax.view_init(elev=-150, azim=70)
ax.legend(diagnoses)
ax.grid()
```

As we can see on the plot similar features are now closer to each other and not so distributed across the plot.
